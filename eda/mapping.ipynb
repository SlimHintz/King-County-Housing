```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

from warnings import filterwarnings

filterwarnings('ignore')

%matplotlib inline 
```

```python
df = pd.read_csv("../raw/kc_house_data_train.csv", index_col=0)
```

```python
# longitude is represented as a floating point number
df.head()
```

```python
df.corr()
```

```python
# Let's get rid of the everything but the price, longitude and latitude for now
geometry = [Point(xy) for xy in zip(df["long"], df["lat"])]
geometry[:5]
```

```python
crs = {'init': 'epsg:4326'}
geo_df = gpd.GeoDataFrame(df,
                         crs=crs,
                         geometry=geometry)
```

```python
geo_df.head()
```

```python
map_df = gpd.read_file("../raw/mapping/kingscounty-shp/Cities_and_Unincorporated_King_County___city_kc_area.shp")
```
